## import the modules

```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

from pandas import *
from datetime import datetime
```

## import the source data file for Cape Town

```python
capetown = read_csv('CapeTown_CPT_2014.csv', skipinitialspace=True)
capetown.head()
```

## find missing values in Events column:

```python
capetown['Events'].isnull()
```

## fill rows with missing col values (NaN) with empty string:

```python
capetown['Events'] = capetown['Events'].fillna('')
capetown['Events'].isnull()
```

## drop rows with missing column values

```python
# capetown.dropna()
```

## check datatypes

```python
capetown.dtypes
```

## set datatypes

```python
capetown = capetown.rename(columns={'WindDirDegrees<br />' : 'WindDirDegrees'}) #rename column
capetown['WindDirDegrees'] = capetown['WindDirDegrees'].str.rstrip('<br />') #strip invalid characters
capetown['WindDirDegrees'] = capetown['WindDirDegrees'].astype('float64') #change datatype to float64
capetown['Date'] = to_datetime(capetown['Date']) #change datatype to datetime
```

## set index to date/time column

```python
capetown.index = capetown['Date']
capetown.head()
```

## for northern hemisphere, set summer to june-august
## for southern hemisphere, set summer to december-february

```python
summer = capetown.loc[datetime(2014,1,1) : datetime(2014,2,28)]
summer[summer['Mean TemperatureC'] >= 25]
```

## find all temperatures over x
## plot a graph of mean temp, over the period

```python
%matplotlib inline
summer['Mean TemperatureC'].plot(grid=True, figsize=(10,5))
```

## plot a graph of mean temp & precipitation, over the period

```python
summer[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

## note best combination of temp/precipitation on graph

```python
feb = summer.loc[datetime(2014,2,14) : datetime(2014,2,28)]
feb[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```
